```python
from nlputils.lexical import Preprocessing
from operator import itemgetter
```

```python
pre = Preprocessing()
```

```python
text = "@vitoriafamer @clarobrasil @prefeitura_poa sim! o preco da vivo e bem salgado mas pelo menos para mim a fatura vem sempre o mesmo valor sem acrescimo adicional"
```

```python
text
```

```python
text = pre.remove_acentos(text)
```

```python
text = pre.tokenize_words(text)
```

```python
text
```

```python
freq = {}
```

```python
for texto in text:
    if texto in freq:
        freq[texto] += 1
    else:
        freq[texto] = 1
        
    
```

```python
freq.items()
```

```python
freq = {}
```

```python
with open ('data/claro/claro-texto-tratado-2019-04-18.txt', 'r') as file:
    lines = file.readlines()
    for line in lines:
        line = pre.remove_acentos(line)
        line = pre.tokenize_words(line)
        for word in line:
            if word in freq:
                freq[word] += 1
            else:
                freq[word] = 1
        
    
    
```

```python
sorted(freq.items(), key=itemgetter(1), reverse=True)
```
